## jupyter使用的快捷键 

只要按enter就是开始编辑，只要按esc就是停止编辑  
只要是ctrl＋enter就是运行或者渲染当前单元格  
如果是shift＋enter就是运行或者渲染当前单元格，并且跳转到下一个单元格  
a就是在当前单元格前面加一个单元格  
b就是在当前单元格后面加一个单元格  
c就是复制当前单元格  
v就是粘贴当前单元格  
m就是把代码块变成md编辑块  
y就是把md编辑快变成代码块  




下面是文本编辑的快捷键
home就是光标移动到本行的最前面  
end就是光标移动到本行的最后面  
ctrl ＋home    到达文章最前面  
ctrl＋end   到达文章最后面  
shift+home  选中光标到本行的最前面  
shift+end   选中光标到本行的最后   
shift+del   删除本行  
ctrl +shift+home就是选中光标到文章的开始  
ctrl+shift+end就是选中光标到文章的末尾
